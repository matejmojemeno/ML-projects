```
import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.linear_model import Ridge, LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeRegressor
import sklearn.metrics as metrics
```

```
data = pd.read_csv('LifeExpectancyData.csv')
```

Z načítaných dát odstránime záznamy bez vysvetlovanej premennej.

```
data.rename(columns =lambda x: x.strip(), inplace=True)
data.dropna(subset=['Life expectancy'], inplace=True)
data.reset_index(drop=True, inplace=True)
```

```
data.head()
```

```
data.info()
```

Počet chýbajúcich hodnôt v každom stĺpci.

```
data.isna().sum()
```

Doplnenie dát pomocou modulu SimpleImputer

```
imputer=SimpleImputer(missing_values=np.nan,strategy='mean',fill_value=None)
 
for column in ['Alcohol', 'Hepatitis B', 'BMI', 'Polio', 'Total expenditure', 'Diphtheria', 'GDP', 'Population', 'thinness  1-19 years', 'thinness 5-9 years', 'Income composition of resources', 'Schooling']:
    data[column] = imputer.fit_transform(data[[column]])
```

Kontrola, čí sú všetky dáta doplnené.

```
data.isna().sum()
```

Graf korelácie medzi stĺpcami DataFramu.

```
corr = data.corr()
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

Využijeme onehotencoding a vytvoríme si **dummies** pre stĺpce country a status.

```
country_dummy = pd.get_dummies(data.Country)
status_dummy = pd.get_dummies(data.Status)

data = pd.concat([data, country_dummy, status_dummy], axis=1)

data.drop(['Country', 'Status'], inplace=True, axis=1)
```

Rozdelenie dát na trénovaciu, validačnú a testovaciu podmnožiunu pomocou funkcie **train_test_split**.

```
random_seed = 42
Xtrain, Xval, ytrain, yval = train_test_split(data.drop(columns = ['Life expectancy']), data['Life expectancy'], test_size=0.4, random_state=random_seed)
Xval, Xtest, yval, ytest = train_test_split(Xval, yval, test_size=0.3, random_state=random_seed)
```

# Lineárna regresia

```
clf = LinearRegression()
clf.fit(Xtrain, ytrain)

# print(clf.coef_)
print(clf.intercept_)

plt.scatter(yval, clf.predict(Xval))
plt.plot([40,90], [40,90], 'r')
plt.show()

print('RMSE:', mean_squared_error(np.array(yval), clf.predict(Xval), squared=False))
print('MAE:', mean_absolute_error(np.array(yval), clf.predict(Xval)))
```

Vytvoril som všetky kombinácie parametrov lineárneho regresoru, a z nich som vybral kombináciu s najlepšími výsledkami.

```
param_grid = {
    'fit_intercept' : [True, False],
    'copy_X' : [True, False],
    'positive' : [True, False]
}

rmse_min = 100
mae_min = 100

for i in param_grid['fit_intercept']:
    for j in param_grid['copy_X']:
        for k in param_grid['positive']:
            clf_tuned = LinearRegression(fit_intercept=i, copy_X=j, positive=k)
            clf_tuned.fit(Xtrain, ytrain)

            rmse = mean_squared_error(np.array(yval), clf_tuned.predict(Xval), squared=False)
            mae = mean_absolute_error(np.array(yval), clf_tuned.predict(Xval))
            if  rmse < rmse_min:
                optimal_rmse_params = [i, j, k]
                rmse_min = rmse
            if mae < mae_min:
                optimal_mae_params = [i, j, k]
                mae_min = mae


clf_rmse_optimal = LinearRegression(fit_intercept=optimal_rmse_params[0],
                                    copy_X=optimal_rmse_params[1],
                                    positive=optimal_rmse_params[2])
clf_rmse_optimal.fit(Xtrain, ytrain)
print('Optimal RMSE:', mean_squared_error(np.array(yval), clf_rmse_optimal.predict(Xval), squared=False))

clf_mae_optimal = LinearRegression(fit_intercept=optimal_mae_params[0],
                                    copy_X=optimal_mae_params[1],
                                    positive=optimal_mae_params[2])
clf_mae_optimal.fit(Xtrain, ytrain)
print('Optimal RMSE:', mean_absolute_error(np.array(yval), clf_mae_optimal.predict(Xval)))
```

# Hrebeňová regresia

```
clf_ridge = Ridge(1)
clf_ridge.fit(Xtrain, ytrain)

plt.scatter(yval, clf_ridge.predict(Xval))
plt.plot([40,90], [40,90], 'r')
plt.show()

print('RMSE:', mean_squared_error(yval, clf_ridge.predict(Xval), squared=False))
print('MAE:', mean_absolute_error(yval, clf_ridge.predict(Xval)))
```

Pomocou modulu **optimize** nájdeme pre hrebeňovú regresiu optimálnu hodnotu alfa.

```
from scipy import optimize

def rmse_eval(alpha):
    clf = Ridge(alpha=alpha)
    clf.fit(Xtrain, ytrain)
    return mean_squared_error(yval, clf.predict(Xval), squared = False)

def mae_eval(alpha):
    clf = Ridge(alpha=alpha)
    clf.fit(Xtrain, ytrain)
    return mean_absolute_error(yval, clf.predict(Xval))

opt_rmse_alpha = optimize.minimize_scalar(rmse_eval, options = {'maxiter': 50}, method = 'bounded', bounds=(0.1, 400))
opt_mae_alpha = optimize.minimize_scalar(mae_eval, options = {'maxiter': 50}, method = 'bounded', bounds=(0.1, 400))

clf_opt_rmse = Ridge(alpha = opt_rmse_alpha.x)
clf_opt_rmse.fit(Xtrain, ytrain)

print('Optimal alpha', opt_rmse_alpha)
print('Optimal RMSE:', mean_squared_error(yval, clf_opt_rmse.predict(Xval), squared=False))

clf_opt_mae = Ridge(alpha = opt_mae_alpha.x)
clf_opt_mae.fit(Xtrain, ytrain)

print('Optimal alpha', opt_mae_alpha)
print('Optimal MAE:', mean_absolute_error(yval, clf_opt_mae.predict(Xval)))
```

# Rozhodovací strom

Ako ďalší model som použil rozhodovací strom, odhadol som optimálnu hĺbku stromu, a model predpovedal dáta s RMSE ~ 2.7

```
rmse_acc = []
mae_acc = []

for depth in range(1, 100):
    clf = DecisionTreeRegressor(max_depth=depth)
    clf.fit(Xtrain, ytrain)
    rmse_acc.append(mean_squared_error(yval, clf.predict(Xval), squared=False))
    mae_acc.append(mean_absolute_error(yval, clf.predict(Xval)))

depth_rmse = np.argmin(rmse_acc) + 1
depth_mae = np.argmin(mae_acc) + 1


clf_rmse = DecisionTreeRegressor(max_depth=depth)
clf_rmse.fit(Xtrain, ytrain)

print('Optimal depth RMSE:', depth_rmse)
print('RMSE:', mean_squared_error(yval, clf_rmse.predict(Xval), squared=False))


clf_mae = DecisionTreeRegressor(max_depth=depth)
clf_mae.fit(Xtrain, ytrain)

print('Optimal depth MAE:', depth_mae)
print('MAE:', mean_absolute_error(yval, clf_mae.predict(Xval)))
```

# Záver

Na záver porovnáme všetky tri modely na novej množine dát, a zistíme, že lineárna a hrebenová regresia vyhodnotia dáta približne rovnako dobre, pričom rozhodovací strom bude zaostávať.

```
print('Linear regression:', mean_squared_error(np.array(ytest), clf_rmse_optimal.predict(Xtest), squared=False))
print('Ridge regression:', mean_squared_error(ytest, clf_opt_rmse.predict(Xtest), squared=False))
print('Decision tree:', mean_squared_error(yval, clf_rmse.predict(Xval), squared=False))
```
